```python
import os
import pandas as pd
import requests
from tqdm import tqdm
df = pd.read_csv("Dataset\\BuzzFeed_fake_news_content.csv",encoding="latin-1")
#print(ifnd.head())
df.describe()
```

```python
# 🔹 Create folder to save images
save_folder = "BF_Images"
os.makedirs(save_folder, exist_ok=True)

# 🔹 Download images
for i, url in tqdm(enumerate(df['top_img'], start=1), total=len(df)):
    try:
        #if i > 56714:
            response = requests.get(url, timeout=10)

            if response.status_code == 200:
                file_path = os.path.join('BF_Images', f"{i}.jpg")

                with open(file_path, "wb") as f:
                    f.write(response.content)
            else:
                print(f"Failed at {i}: Status {response.status_code}")

    except Exception as e:
        print(f"Error at {i}: {e}")
```
