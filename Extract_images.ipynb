```python
import os
import pandas as pd
import requests
from tqdm import tqdm
df = pd.read_csv("Dataset\\ifnd.csv",encoding="latin-1")
#print(ifnd.head())
df.describe()
```

```python
# 🔹 Create folder to save images
# save_folder = "BF_Images"
# os.makedirs(save_folder, exist_ok=True)

# 🔹 Download images
for i, url in tqdm(enumerate(df['Image'], start=1), total=len(df)):
    try:
        if i > 25756:
            response = requests.get(url, timeout=10)
            failed_images=[]
            # if response.status_code != 200:
            #     failed_images.append(i)

            if response.status_code == 200:
                file_path = os.path.join('Images', f"{i}.jpg")

                with open(file_path, "wb") as f:
                    f.write(response.content)
            else:
                print(f"Failed at {i}: Status {response.status_code}")

    except Exception as e:
        print(f"Error at {i}: {e}")
```
